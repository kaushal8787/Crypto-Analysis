```python
#         :
# pip install plotly pandas

import pandas as pd
import plotly.express as px

# --- 1.  ---
#       (USD to INR)
USD_TO_INR_RATE = 88.02145 
file_name = "USDC-USD From 2018 To Dec-2024.csv"

# --- 2.     ---

#  :  'utf-8'   ,  'latin-1'   
try:
    df = pd.read_csv(file_name, encoding='utf-8') 
except UnicodeDecodeError:
    df = pd.read_csv(file_name, encoding='latin-1')
    
# 'Date'   datetime   
df['Date'] = pd.to_datetime(df['Date'])

# --- 3.   (INR Conversion) ---

# USD  Open  Close   INR  
df['Open Price (INR)'] = df['Open'] * USD_TO_INR_RATE
df['Close Price (INR)'] = df['Close'] * USD_TO_INR_RATE

# --- 4.    (Plotly Express) ---

fig = px.line(df,
              x='Date',
              y='Close Price (INR)',
              title='USDC-USD Closing Price Over Time (in Indian Rupees - Approx.)',
              # Hover_data (Tooltip)      :
              hover_data={'Date': '|%Y-%m-%d',
                          'Open Price (INR)': ':.2f',
                          'Close Price (INR)': ':.2f',
                          'Close': False}
             )

# --- 5. -/     ---

# X-axis    
fig.update_xaxes(
    rangeslider_visible=True,
    #    
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all", label="All")
        ])
    )
)

#      
fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Closing Price (INR)",
    hovermode="x unified",
    #   10x6    1000x600    
    width=800,
    height=600
)

#      
fig.show()
```

```python
#         :
# pip install plotly pandas

import pandas as pd
import plotly.graph_objects as go
import plotly.express as px # Rangeslider setup  

# --- 1.     ---
#       (USD to INR)
USD_TO_INR_RATE = 88.02145 
file_name = "USDC-USD From 2018 To Dec-2024.csv"

#  : 'latin-1'    
try:
    df = pd.read_csv(file_name, encoding='latin-1')
except Exception:
    print("Warning: Could not read file with 'latin-1' encoding. Using default.")
    df = pd.read_csv(file_name)

df['Date'] = pd.to_datetime(df['Date'])

# --- 2. INR  ---

# Candlestick chart   Open, High, Low, Close  INR  
df['Open Price (INR)'] = df['Open'] * USD_TO_INR_RATE
df['High Price (INR)'] = df['High'] * USD_TO_INR_RATE
df['Low Price (INR)'] = df['Low'] * USD_TO_INR_RATE
df['Close Price (INR)'] = df['Close'] * USD_TO_INR_RATE

# --- 3. Candlestick   ---

fig = go.Figure(data=[go.Candlestick(
    x=df['Date'],
    open=df['Open Price (INR)'],
    high=df['High Price (INR)'],
    low=df['Low Price (INR)'],
    close=df['Close Price (INR)'],
    name='USDC/INR Price Action'
)])

# --- 4.   -   ---

# Candlestick chart   ,  rangeslider_visible   x-axis     
fig.update_layout(
    title='USDC Candlestick Chart (Open, High, Low, Close) in Indian Rupees',
    xaxis_title='Date',
    yaxis_title='Price (INR)',
    width=800, #  10   
    height=600, #  6   
    # -/  
    xaxis_rangeslider_visible=True,
    xaxis_rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all", label="All")
        ])
    )
)

#                
fig.update_yaxes(
    fixedrange=False # Y-axis      
)

#   
fig.show()
```

```python
#         :
# pip install plotly pandas numpy

import pandas as pd
import plotly.express as px
import numpy as np

# --- 1.     ---
file_name = "USDC-USD From 2018 To Dec-2024.csv"

#  : 'latin-1'    
try:
    df = pd.read_csv(file_name, encoding='latin-1')
except Exception:
    print("Warning: Could not read file with 'latin-1' encoding. Using default.")
    df = pd.read_csv(file_name)

# --- 2. Log Return     ---

# Log Return        
df['Log Return'] = pd.to_numeric(df['Log Return'], errors='coerce')
df.replace([np.inf, -np.inf], np.nan, inplace=True)
# NaN        
df.dropna(subset=['Log Return'], inplace=True) 

# Log Return   (%)  
df['Log Return (%)'] = df['Log Return'] * 100

# --- 3.    ---

# Plotly Express      KDE (Kernel Density Estimate -  ) 
fig = px.histogram(df, 
                   x='Log Return (%)',
                   nbins=80,  #    (       )
                   marginal="box", #     
                   histnorm='probability density', # Y-axis   (Density) 
                   title='Distribution of USDC Daily Log Returns (in Percentage)')

# --- 4.   ---

fig.update_layout(
    xaxis_title="Daily Log Return (%)",
    yaxis_title="Density (Frequency)",
    hovermode="x unified",
    width=1000, #  10   
    height=600, #  6   
    bargap=0.01 #      
)

#        (KDE)  :
fig.update_traces(
    # KDE    (     )
    selector=dict(type='histogram'), 
    # KDE     marginal="box"   marginal="kde"   
    #            
)

#   
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# --- 1.     ---
#       (USD to INR)
USD_TO_INR_RATE = 88.02145 
file_name = "USDC-USD From 2018 To Dec-2024.csv"

#  : 'latin-1'    
try:
    df = pd.read_csv(file_name, encoding='latin-1')
except Exception:
    print("Warning: Could not read file with 'latin-1' encoding. Using default.")
    df = pd.read_csv(file_name)

df['Date'] = pd.to_datetime(df['Date'])

# --- 2. INR  ---
df['Close Price (INR)'] = df['Close'] * USD_TO_INR_RATE
df['Trading Range (INR)'] = df['Trading Range'] * USD_TO_INR_RATE


# --- 3.   1: Price (INR) vs. Volume (Close Price  ) ---
fig1 = px.scatter(df, 
                  x='Volume', 
                  y='Close Price (INR)', 
                  color='Close Price (INR)', # <--- Color based on Close Price
                  color_continuous_scale=px.colors.sequential.Viridis, #  
                  title='1. USDC Close Price (INR) vs. Volume (Colored by Price)',
                  hover_data={'Date': True, 'Close Price (INR)': ':.2f', 'Volume': ':.0f'}) 

fig1.update_layout(
    xaxis_title='Volume (Units of USDC)',
    yaxis_title='Close Price (INR)',
    hovermode="closest",
    width=1000,
    height=600
)
fig1.show()


# --- 4.   2: Trading Range (INR) vs. Volume (Close Price  ) ---
fig2 = px.scatter(df, 
                  x='Volume', 
                  y='Trading Range (INR)', 
                  color='Close Price (INR)', # <--- Color based on Close Price
                  color_continuous_scale=px.colors.sequential.Plasma, #   
                  title='2. USDC Trading Range (INR) vs. Volume (Colored by Price)',
                  hover_data={'Date': True, 'Trading Range (INR)': ':.2f', 'Volume': ':.0f'}) 

fig2.update_layout(
    xaxis_title='Volume (Units of USDC)',
    yaxis_title='Trading Range (INR) - (High Price - Low Price)',
    hovermode="closest",
    width=1000,
    height=600
)
fig2.show()
```

```python
#         :
# pip install plotly pandas numpy

import pandas as pd
import plotly.express as px
import numpy as np

# --- 1.     ---
file_name = "USDC-USD From 2018 To Dec-2024.csv"
WINDOW = 30 # 30-   
ANNUALIZATION_FACTOR = np.sqrt(252) #     (252  )

#  : 'latin-1'    
try:
    df = pd.read_csv(file_name, encoding='latin-1')
except Exception:
    print("Warning: Could not read file with 'latin-1' encoding. Using default.")
    df = pd.read_csv(file_name) 

df['Date'] = pd.to_datetime(df['Date'])

# --- 2. Log Return     ---
df['Log Return'] = pd.to_numeric(df['Log Return'], errors='coerce')
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(subset=['Log Return'], inplace=True)

# --- 3.   (Rolling Volatility)   ---
# 30-     (Standard Deviation)  
df['Rolling Volatility (Daily)'] = df['Log Return'].rolling(window=WINDOW).std()

#   (Annualized Volatility)      
df['Annualized Volatility (%)'] = df['Rolling Volatility (Daily)'] * ANNUALIZATION_FACTOR * 100

# --- 4.   ---
fig = px.line(df,
              x='Date',
              y='Annualized Volatility (%)',
              title=f'USDC Annualized Rolling Volatility ({WINDOW}-Day Window)')

# --- 5.   -   ---
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all", label="All")
        ])
    )
)

fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Annualized Volatility (%)",
    hovermode="x unified",
    width=1000,
    height=600
)

#   
fig.show()
```

```python
#          :
# !pip install pandas plotly statsmodels

import pandas as pd
import plotly.express as px

# --- 1.     ---
USD_TO_INR_RATE = 88.02145 #   
file_name = "USDC-USD From 2018 To Dec-2024.csv"

#  : 'latin-1'    
try:
    df = pd.read_csv(file_name, encoding='latin-1')
except Exception:
    print("Warning: Could not read file with 'latin-1' encoding.")
    df = pd.read_csv(file_name) 

# 'Date'   datetime   
df['Date'] = pd.to_datetime(df['Date'])

# --- 2.         INR   ---

#  (Monday)      (dayofweek 0 = Monday)
df_monday = df[df['Date'].dt.dayofweek == 0].copy()

# INR     
df_monday['Open Price (INR)'] = df_monday['Open'] * USD_TO_INR_RATE
df_monday['Close Price (INR)'] = df_monday['Close'] * USD_TO_INR_RATE

# --- 3.   (Correlation Chart)  ---

fig = px.scatter(df_monday, 
                 x='Open Price (INR)', 
                 y='Close Price (INR)', 
                 trendline='ols', # OLS (Ordinary Least Squares)  
                 title='Correlation between Opening and Closing Prices on MONDAYS (in INR)',
                 hover_data={'Date': True}) 

# --- 4.     ---
fig.update_layout(
    xaxis_title='Opening Price (INR)',
    yaxis_title='Closing Price (INR)',
    width=1000,
    height=600
)

#   
fig.show()
```
